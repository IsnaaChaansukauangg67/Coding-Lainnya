```python
#menghitung nilai statistik
import statistics as stats

#operasi numerik dan perhitungan matematis
import numpy

#u/ menampilakan grafik atau visualisasi data
import matplotlib.pyplot as plt
```

```python
#data penjualan
penjualan = {250,300,260,270,275,290,310,320,330,335,
             340,350,355,360,370,380,385,400,405,410,
             420,430,435,440,450,470,480,490,500,550}
```

```python

mean = stats.mean (penjualan)
median = stats.median(penjualan)
mode = stats.mode(penjualan)
stdev = stats.stdev(penjualan)
var = stats.variance(penjualan)
cv = (stdev / mean)*100
```

```python
#menampilkan hasil perhitungan statistik di layar dg pembulatan 2 angka blkg koma
print("Rata- rata (Mean):", mean)
print("Median           :", median)
print("Modus            :", mode)
print("Standar Deviasi  :",round(stdev,2))
print("Variasi          :",round(var,2))
print("Koefisien Variasi:",round(cv,2),"%")

#menampilkan grafik np.histogram
plt.hist(penjualan,bins=8,edgecolor= 'black')
plt.title("Distribusi Penjualan Harian (Ribu Rupiah)")
plt.xlabel("Penjualan (ribu)")
plt.ylabel("Frekuensi")
plt.show()
```
